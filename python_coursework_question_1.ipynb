```python
import pandas as pd
```

```python
loan=pd.read_csv('loan50.csv')
```

```python
ir = ['interest_rate']
```

```python
mean = loan['interest_rate'].mean()
```

```python
median = loan['interest_rate'].median()
```

```python
std = loan['interest_rate'].std()
```

```python
q1 = loan['interest_rate'].quantile(0.25)
```

```python
q3 = loan['interest_rate'].quantile(0.75)
```

```python
ik iqr = q3 - q1
```

```python
print('Mean:', mean)
print('Median:', median)
print('Standard Deviation:', std)
print('First Quartile (Q1):', q1)
print('Third Quartile (Q3):', q3)
print('Interquartile Range (IQR):', iqr)
```

```python
table1 = pd.DataFrame({
    'Measure': ['Mean', 'Median', 'Standard Deviation', 'First Quartile (Q1)', 'Third Quartile (Q3)', 'Interquartile Range (IQR)'],
    'Value': [mean, median, std, q1, q3, iqr]})
```

question 2c

because the distribution is right-skewed, the median is a more appropriate measure of centre. This is because the mean would be pulled by the few high-rate loans, but the median would not be pulled, just showing the middle of the data.

question 3b

```python
loan.boxplot(column='interest_rate', by='homeownership')
plt.xlabel('Homeownership Status')
plt.ylabel('Interest Rate')
plt.title('Interest Rate by Homeownership Status')
plt.show()
```

 mortgage borrowers have the widest spread and highest maximum, with a median around 9.3%. The distribution has a long upper tail.
Renters have a tighter IQR (7.96-10.91%), a similar median to mortgage holders (around 9.92%) but fewer extreme high-rate outliers. The distribution is more compact and right-skewed.
Owners have the highest median (around 15%), and their entire box sits higher than the other two groups, however there are only 3 observations so reliable conclusions cannot be drawn.


question 3c

A bar chart is used when the variable is categorical, such as homeownership (mortgage, rent, own). Each bar is a separate group, and the order of the bars is arbitrary.
A histogram is used when the variable is numerical and continuous, such as interest rate. The x-axis is a true number line, bars represent intervals, and the overall shape of the bars tells you about if it's skewed, the modality and the spread. Gaps between the bars could be misleading because the scale is continuous.
